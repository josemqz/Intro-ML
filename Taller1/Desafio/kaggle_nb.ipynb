```python
def lemmatizeAndStem(phrase):
    lemmList = [simplemma.lemmatize(i, langdata) for i in phrase.split(" ") if i not in string.punctuation and not i.isdigit()]
    stemList = [ita_stemmer.stem(i) for i in lemmList]
    LaSstring = " ".join(stemList)
    return LaSstring
```

```python
tfidf1 = TfidfVectorizer(smooth_idf=False,sublinear_tf=True, min_df=4, norm='l2', encoding='utf-8', ngram_range=(1, 3), stop_words=stop_words_it)
tfidf2 = TfidfVectorizer(smooth_idf=False,sublinear_tf=True, min_df=4, norm='l2', encoding='utf-8', ngram_range=(1, 3), stop_words=stop_words_it)
tfidf1_fit = tfidf1.fit(df['description'])
tfidf2_fit = tfidf2.fit(df['title'])

features_desc = tfidf1_fit.transform(df['description']).toarray()
features_tit = tfidf2_fit.transform(df['title']).toarray()
feat_full = np.hstack((features_desc, features_tit))
```

```python
#cross-validation
global_loss = 0.0
cant_fold = 5
kfolder = KFold(n_splits=cant_fold, shuffle=True)
for train_index, test_index in kfolder.split(feat_full):
    train_df = feat_full[train_index]
    test_df = feat_full[test_index]
    mnb_it = MultinomialNB().fit(train_df, df.iloc[train_index]['class'])
    predicciones = mnb_it.predict(test_df)
    clases_loc = df['class'].iloc[test_index].to_numpy()
    resutlados = predicciones == clases_loc
    global_loss += len(resutlados) - resutlados[resutlados].sum()
print(global_loss/len(df))
```

```python
#modelo
mnb_it = MultinomialNB().fit(feat_full, df['class'])
#predicciones

name = 'ferramenta_competition_test_without_labels.csv'
df2 = pd.read_csv(path+name)
df2['title'] = df2['title'].map(str.lower)
df2['description'] = df2['description'].map(str.lower)
df2['title'] = df2['title'].apply(lemmatizeAndStem)
df2['description'] = df2['description'].apply(lemmatizeAndStem)

features_desc2 = tfidf1_fit.transform(df2['description']).toarray()
features_tit2 = tfidf2_fit.transform(df2['title']).toarray()
#features_img = df.loc[:, 'f0':'f255'].to_numpy()
feat_full2 = np.hstack((features_desc2, features_tit2))

predicciones = mnb_it.predict(feat_full2)

with open('resulta3.csv', 'w') as arc:
    arc.write("image,class\n")
    for i, clase in enumerate(predicciones):
        arc.write(f"{df2['image'].iloc[i]},{clase}\n")
```
